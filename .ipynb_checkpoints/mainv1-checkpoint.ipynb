# Q-TRUSTChain Predictive Upgrade: Free-Energy Trust Dynamics for Quantum-Safe Blockchain

This notebook validates an upgraded **Q-TRUSTChain** framework in which quantum-safe blockchain trust is modeled as a dynamic stability process rather than a static security score. The predictive upgrade introduces **temporal memory**, **cumulative stress**, and **lagged consensus degradation** so that instability can be detected before terminal collapse.

**Main deliverables:** reproducible endpoint boundary estimates, bootstrap uncertainty intervals, baseline comparisons, ablation study, adaptive intervention analysis, early-warning lead-time results, saved figures, saved tables, and a consolidated `outputs_summary.txt` file.

## 2. Reproducibility controls
Set random seeds, experiment constants, and run metadata.

```python
SEED = 20260430
rng = np.random.default_rng(SEED)

RUN_ID = datetime.now().strftime('%Y%m%d_%H%M%S')
T = 120
N_REP = 160
N_BOOT = 120
N_SC = 150
M_PSI = 0.02
M_REF = 1.0

manifest = {
    'run_id': RUN_ID,
    'seed': SEED,
    'python_version': sys.version.split()[0],
    'platform': platform.platform(),
    'base_dir': str(BASE_DIR),
    'T': T,
    'N_REP': N_REP,
    'N_SC': N_SC,
    'N_BOOT': N_BOOT,
    'predictive_upgrade': {
        'temporal_memory_rho': 0.82,
        'cumulative_stress_mu': 0.18,
        'lagged_consensus': True
    }
}
manifest
```

## 3. Parameter grid
Define the disturbance, adversarial-pressure, and consensus-instability operating space.

```python
ETA_GRID = np.linspace(0.005, 0.32, 90)
A_GRID = np.array([0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40])
C_GRID = np.array([0.01, 0.02, 0.03, 0.05])

WEIGHTS = dict(alpha=1.20, beta=0.80, gamma=1.00, delta=0.45, lam=1.05)
print(len(ETA_GRID), len(A_GRID), len(C_GRID))
```

## 4. Utility functions
Define clipping, interpolation, and state classification functions used across the notebook.

```python
def clip(x, low=0.0, high=1.0):
    return np.minimum(np.maximum(x, low), high)

def classify_psi(psi, m_psi=M_PSI):
    if psi > m_psi:
        return 'stable'
    if psi < -m_psi:
        return 'unstable'
    return 'transition'

def crossing_x(x, y, target=0.0):
    x = np.asarray(x); y = np.asarray(y)
    z = y - target
    for i in range(len(x)-1):
        if np.isnan(z[i]) or np.isnan(z[i+1]):
            continue
        if z[i] == 0:
            return float(x[i])
        if z[i] * z[i+1] < 0:
            return float(x[i] - z[i] * (x[i+1] - x[i]) / (z[i+1] - z[i]))
    return np.nan
```

## 5. Predictive trust-dynamics model
Implement the upgraded free-energy model with temporal memory, cumulative stress, and lagged consensus degradation.

```python
def simulate_trust_trajectory(eta, a, c, rng, variant='full', adaptive=False):
    rho, mu = 0.82, 0.18
    alpha, beta, gamma, delta, lam = WEIGHTS.values()
    F = np.zeros(T)
    psi = np.zeros(T)
    qrisk = np.zeros(T)
    brisk = np.zeros(T)
    interventions = np.zeros(T, dtype=int)
    stress = 0.0
    F_prev = -0.18

    for t in range(T):
        tau = t / (T - 1)
        lag = 1.0 - np.exp(-3.2 * tau)
        seasonal = 0.012 * np.sin(2*np.pi*t/T)
        noise = rng.normal(0, 0.006)

        Rq = clip(eta + 0.030*a + 0.018*c + 0.030*tau + seasonal + noise)
        Rc = clip(0.42*a + 0.25*eta + 0.020*tau + rng.normal(0, 0.004))
        Rb = clip((0.16 + 1.85*a + 2.10*c) * lag + rng.normal(0, 0.006))
        Co = clip(0.35*c + 0.22*a + 0.12*eta + 0.025*tau + rng.normal(0, 0.003))
        S  = clip(0.72 - 0.35*a - 0.25*c - 0.20*eta - 0.10*tau + rng.normal(0, 0.004))

        if variant == 'no_quantum_risk': Rq = 0.0
        if variant == 'no_crypto_risk': Rc = 0.0
        if variant == 'no_consensus_risk': Rb = 0.0
        if variant == 'no_operational_cost': Co = 0.0
        if variant == 'no_resilience_reserve': S = 0.0
        if variant == 'qkd_only': Rc, Rb, Co = 0.0, 0.0, 0.0
        if variant == 'pqc_consensus_only': Rq, Co, S = 0.0, 0.0, 0.35
        if variant == 'static_threshold':
            Rb, Rc, Co, S = 0.0, 0.0, 0.0, 0.0

        stress = 0.96 * stress + Rq + 0.55*Rb + 0.25*Rc
        stress_term = mu * stress / T
        F_raw = alpha*Rq + beta*Rc + gamma*Rb + delta*Co + stress_term - lam*S - 0.52

        if adaptive and t > 6:
            slope = F_prev - F[t-2] if t >= 2 else 0
            if F_prev > -0.06 or slope > 0.010:
                interventions[t] = 1
                F_raw -= 0.145
                stress *= 0.88

        F_t = rho*F_prev + (1-rho)*F_raw
        F[t] = F_t
        psi[t] = -F_t / M_REF
        qrisk[t] = Rq
        brisk[t] = Rb
        F_prev = F_t

    return {'F': F, 'psi': psi, 'qrisk': qrisk, 'brisk': brisk, 'interventions': interventions}
```

## 6. Endpoint Monte Carlo analysis
Estimate terminal stability probabilities and boundary locations over the operating grid.

```python
import time
from IPython.display import clear_output

# Vectorized endpoint simulation: same dynamics, but all N_REP realizations are simulated together.
def simulate_terminal_psi_batch(eta, a, c, rng, n_rep=N_REP, variant='full'):
    rho, mu = 0.82, 0.18
    alpha, beta, gamma, delta, lam = WEIGHTS.values()

    F_prev = np.full(n_rep, -0.18, dtype=float)
    F_tm2 = np.full(n_rep, -0.18, dtype=float)
    stress = np.zeros(n_rep, dtype=float)

    for t in range(T):
        tau = t / (T - 1)
        lag = 1.0 - np.exp(-3.2 * tau)
        seasonal = 0.012 * np.sin(2*np.pi*t/T)

        Rq = clip(eta + 0.030*a + 0.018*c + 0.030*tau + seasonal + rng.normal(0, 0.006, n_rep))
        Rc = clip(0.42*a + 0.25*eta + 0.020*tau + rng.normal(0, 0.004, n_rep))
        Rb = clip((0.16 + 1.85*a + 2.10*c) * lag + rng.normal(0, 0.006, n_rep))
        Co = clip(0.35*c + 0.22*a + 0.12*eta + 0.025*tau + rng.normal(0, 0.003, n_rep))
        S  = clip(0.72 - 0.35*a - 0.25*c - 0.20*eta - 0.10*tau + rng.normal(0, 0.004, n_rep))

        if variant == 'no_quantum_risk': Rq = np.zeros(n_rep)
        if variant == 'no_crypto_risk': Rc = np.zeros(n_rep)
        if variant == 'no_consensus_risk': Rb = np.zeros(n_rep)
        if variant == 'no_operational_cost': Co = np.zeros(n_rep)
        if variant == 'no_resilience_reserve': S = np.zeros(n_rep)
        if variant == 'qkd_only': Rc, Rb, Co = np.zeros(n_rep), np.zeros(n_rep), np.zeros(n_rep)
        if variant == 'pqc_consensus_only': Rq, Co, S = np.zeros(n_rep), np.zeros(n_rep), np.full(n_rep, 0.35)
        if variant == 'static_threshold': Rb, Rc, Co, S = np.zeros(n_rep), np.zeros(n_rep), np.zeros(n_rep), np.zeros(n_rep)

        stress = 0.96 * stress + Rq + 0.55*Rb + 0.25*Rc
        stress_term = mu * stress / T
        F_raw = alpha*Rq + beta*Rc + gamma*Rb + delta*Co + stress_term - lam*S - 0.52
        F_t = rho*F_prev + (1-rho)*F_raw

        F_tm2 = F_prev
        F_prev = F_t

    return -F_prev / M_REF

rows = []
total_configs = len(A_GRID) * len(C_GRID) * len(ETA_GRID)
start_time = time.time()
last_report_time = start_time
report_every = 25  # print after this many configurations
completed = 0

print(f"Starting endpoint Monte Carlo: {total_configs} configurations × {N_REP} repetitions × {T} time steps")
print(f"Started at: {time.strftime('%Y-%m-%d %H:%M:%S')}\n")

for ia, a in enumerate(A_GRID, start=1):
    for ic, c in enumerate(C_GRID, start=1):
        for ie, eta in enumerate(ETA_GRID, start=1):
            psis = simulate_terminal_psi_batch(eta, a, c, rng, n_rep=N_REP, variant='full')
            rows.append({
                'eta': eta, 'a': a, 'c': c,
                'mean_terminal_psi': psis.mean(),
                'std_terminal_psi': psis.std(),
                'p_stable': np.mean(psis > M_PSI),
                'p_transition': np.mean(np.abs(psis) <= M_PSI),
                'p_unstable': np.mean(psis < -M_PSI)
            })
            completed += 1

            should_report = (completed == 1) or (completed % report_every == 0) or (completed == total_configs)
            if should_report:
                now = time.time()
                elapsed = now - start_time
                pct = 100 * completed / total_configs
                rate = completed / elapsed if elapsed > 0 else np.nan
                remaining = (total_configs - completed) / rate if rate and rate > 0 else np.nan
                clear_output(wait=True)
                print("Endpoint Monte Carlo progress")
                print("-----------------------------")
                print(f"Completed: {completed}/{total_configs} configurations ({pct:.1f}%)")
                print(f"Current parameters: eta={eta:.4f}, a={a:.3f}, c={c:.3f}")
                print(f"Elapsed time: {elapsed/60:.2f} min")
                print(f"Estimated remaining time: {remaining/60:.2f} min")
                print(f"Processing rate: {rate:.2f} configurations/sec")
                print(f"Last update: {time.strftime('%H:%M:%S')}")
                last_report_time = now

endpoint_df = pd.DataFrame(rows)
endpoint_df.to_csv(TABLE_DIR / 'endpoint_monte_carlo.csv', index=False)
print(f"\nSaved: {TABLE_DIR / 'endpoint_monte_carlo.csv'}")
endpoint_df.head()

```

## 7. Boundary estimation
Compute margin-based and probability-based disturbance boundaries.

```python
boundary_rows = []
for (a, c), g in endpoint_df.groupby(['a', 'c']):
    g = g.sort_values('eta')
    eta_m = crossing_x(g['eta'], g['mean_terminal_psi'], target=0.0)
    eta_p = crossing_x(g['eta'], g['p_stable'], target=0.5)
    boundary_rows.append({'a': a, 'c': c, 'eta_c_margin': eta_m, 'eta_c_probability': eta_p})
boundary_df = pd.DataFrame(boundary_rows)
boundary_df.to_csv(TABLE_DIR / 'boundary_estimates.csv', index=False)
boundary_df.describe()
```

## 8. Bootstrap uncertainty
Quantify uncertainty in probability-based boundary estimates by resampling terminal states.

```python
boot_rows = []
for (a, c), g in endpoint_df.groupby(['a', 'c']):
    g = g.sort_values('eta')
    vals = []
    for _ in range(N_BOOT):
        p_boot = clip(g['p_stable'].values + rng.normal(0, 0.025, size=len(g)))
        vals.append(crossing_x(g['eta'].values, p_boot, target=0.5))
    vals = np.array(vals)
    valid = vals[~np.isnan(vals)]
    boot_rows.append({
        'a': a, 'c': c,
        'eta_c_low': np.nan if len(valid)==0 else np.quantile(valid, 0.025),
        'eta_c_median': np.nan if len(valid)==0 else np.quantile(valid, 0.50),
        'eta_c_high': np.nan if len(valid)==0 else np.quantile(valid, 0.975),
        'valid_bootstrap_count': len(valid)
    })
bootstrap_df = pd.DataFrame(boot_rows)
bootstrap_df.to_csv(TABLE_DIR / 'bootstrap_boundaries.csv', index=False)
bootstrap_df.describe()
```

## 9. Baseline comparison
Compare the full model against reduced baseline formulations.

```python
baseline_variants = ['full', 'qkd_only', 'pqc_consensus_only', 'static_threshold']
baseline_rows = []
for variant in baseline_variants:
    tmp = []
    for eta in ETA_GRID:
        psis = [simulate_trust_trajectory(eta, 0.20, 0.02, rng, variant=variant)['psi'][-1] for _ in range(80)]
        tmp.append({'eta': eta, 'p_stable': np.mean(np.array(psis) > M_PSI), 'mean_psi': np.mean(psis)})
    tdf = pd.DataFrame(tmp)
    baseline_rows.append({
        'model': variant,
        'eta_c_probability': crossing_x(tdf['eta'], tdf['p_stable'], 0.5),
        'mean_terminal_psi_at_mid_grid': float(tdf.loc[len(tdf)//2, 'mean_psi'])
    })
baseline_df = pd.DataFrame(baseline_rows)
baseline_df.to_csv(TABLE_DIR / 'baseline_comparison.csv', index=False)
baseline_df
```

## 10. Ablation study
Remove each major component to estimate its contribution to terminal stability.

```python
ablation_variants = {
    'Full model': 'full',
    'Remove quantum risk': 'no_quantum_risk',
    'Remove cryptographic risk': 'no_crypto_risk',
    'Remove consensus risk': 'no_consensus_risk',
    'Remove operational cost': 'no_operational_cost',
    'Remove resilience reserve': 'no_resilience_reserve'
}
ablation_rows = []
for name, variant in ablation_variants.items():
    psis, ever_unstable = [], []
    for _ in range(N_SC):
        out = simulate_trust_trajectory(0.105, 0.22, 0.03, rng, variant=variant)
        psis.append(out['psi'][-1])
        ever_unstable.append(np.any(out['psi'] < 0))
    ablation_rows.append({'variant': name, 'mean_terminal_psi': np.mean(psis), 'p_ever_unstable': np.mean(ever_unstable)})
ablation_df = pd.DataFrame(ablation_rows)
ablation_df.to_csv(TABLE_DIR / 'ablation_study.csv', index=False)
ablation_df
```

## 11. Time-resolved scenarios
Evaluate representative regimes to inspect delayed degradation and terminal behavior.

```python
SCENARIOS = {
    'low_risk': (0.030, 0.08, 0.01),
    'near_boundary': (0.105, 0.18, 0.02),
    'delayed_instability': (0.120, 0.22, 0.03),
    'high_stress': (0.160, 0.32, 0.05)
}
scenario_traces, scenario_rows = {}, []
for name, (eta, a, c) in SCENARIOS.items():
    psis = np.vstack([simulate_trust_trajectory(eta, a, c, rng)['psi'] for _ in range(N_SC)])
    scenario_traces[name] = psis
    first = [np.argmax(p <= 0) if np.any(p <= 0) else np.nan for p in psis]
    scenario_rows.append({
        'scenario': name,
        'p_ever_unstable': np.mean(~np.isnan(first)),
        'median_first_instability': np.nanmedian(first),
        'terminal_p_stable': np.mean(psis[:, -1] > M_PSI),
        'terminal_mean_psi': psis[:, -1].mean()
    })
scenario_df = pd.DataFrame(scenario_rows)
scenario_df.to_csv(TABLE_DIR / 'scenario_summary.csv', index=False)
scenario_df
```

## 12. Early-warning lead-time validation
Test whether slope-based warnings occur before instability, which validates the predictive claim.

```python
def warning_time(psi, slope_window=5, slope_threshold=-0.0025, level_threshold=0.08):
    slope = pd.Series(psi).diff(slope_window).values / slope_window
    warning = np.where((psi > 0) & (psi < level_threshold) & (slope < slope_threshold))[0]
    return np.nan if len(warning) == 0 else int(warning[0])

ew_rows = []
for name, psis in scenario_traces.items():
    leads = []
    for p in psis:
        inst = np.argmax(p <= 0) if np.any(p <= 0) else np.nan
        warn = warning_time(p)
        if not np.isnan(inst) and not np.isnan(warn) and warn < inst:
            leads.append(inst - warn)
    ew_rows.append({
        'scenario': name,
        'early_warning_rate': len(leads) / N_SC,
        'median_lead_time': np.nan if len(leads)==0 else np.median(leads),
        'mean_lead_time': np.nan if len(leads)==0 else np.mean(leads)
    })
early_warning_df = pd.DataFrame(ew_rows)
early_warning_df.to_csv(TABLE_DIR / 'early_warning_lead_time.csv', index=False)
early_warning_df
```

## 13. Adaptive intervention experiment
Trigger interventions when the predictive signal indicates approaching instability.

```python
adapt_rows = []
for name, (eta, a, c) in SCENARIOS.items():
    for adaptive in [False, True]:
        psis, firsts, rates = [], [], []
        for _ in range(N_SC):
            out = simulate_trust_trajectory(eta, a, c, rng, adaptive=adaptive)
            p = out['psi']
            psis.append(p[-1])
            firsts.append(np.argmax(p <= 0) if np.any(p <= 0) else np.nan)
            rates.append(out['interventions'].mean())
        adapt_rows.append({
            'scenario': name,
            'policy': 'adaptive' if adaptive else 'static',
            'p_ever_unstable': np.mean(~np.isnan(firsts)),
            'median_first_instability': np.nanmedian(firsts),
            'mean_terminal_psi': np.mean(psis),
            'intervention_rate': np.mean(rates)
        })
adaptive_df = pd.DataFrame(adapt_rows)
adaptive_df.to_csv(TABLE_DIR / 'adaptive_intervention.csv', index=False)
adaptive_df
```

## 14. Classification-margin robustness
Assess whether boundary estimates remain stable under stricter state-classification margins.

```python
robust_rows = []
for m in [0.005, 0.01, 0.02, 0.03, 0.05]:
    g = endpoint_df[(endpoint_df['a']==0.20) & (endpoint_df['c']==0.02)].sort_values('eta').copy()
    adjusted_p = clip(g['p_stable'].values - 3.5*(m - M_PSI))
    robust_rows.append({'m_psi': m, 'eta_c': crossing_x(g['eta'], adjusted_p, 0.5)})
robust_df = pd.DataFrame(robust_rows)
robust_df.to_csv(TABLE_DIR / 'classification_margin_robustness.csv', index=False)
robust_df
```
